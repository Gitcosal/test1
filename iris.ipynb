```python
# 라이브러리 사용

import tensorflow as tf
import pandas as pd
```

```python
# 데이터 준비

url = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris = pd.read_csv(url)
iris
```

```python
# 원핫인코딩 : 숫자가 아닌 데이터를 0과 1로 나누어주는 것

iris= pd.get_dummies(iris)
iris
```

```python
# 독립변수와 종속변수 분리

독립 = iris[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속 = iris[['품종_setosa', '품종_versicolor', '품종_virginica']]
```

```python
독립
```

```python
종속
```

```python
print(독립.shape, 종속.shape)
```

```python
# 모델 만들기

x = tf.keras.layers.Input(shape=[4])

# softmax : 분류에서 이것일 확률이 몇 퍼센트이다를 나타내는 비율을 예측하는 것
y = tf.keras.layers.Dense(3, activation='softmax')(x)

model = tf.keras.models.Model(x, y)

# crossentropy : 분류에 사용되는 loss
# accuracy : 정확도
model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# 모델 학습

model.fit(독립, 종속, epochs=1000, verbose=0)
model.fit(독립, 종속, epochs=10)
```

```python
model.predict(독립[:5])
```
